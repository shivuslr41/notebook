```python
import pandas as pd
import numpy as np
```

```python
questions = ["who is the prime minister of india",
             "who is the president of india",
             "34513",
             "43155",
             "52446",
             "654512",
             "73512",
             "83515",
             "95624",
             "10464",
             ]
```

```python
questions_options = {questions[0]: ["Modi", "Rahul", "Trump", "None of the above"],
                    questions[1]: ["Modi", "Rahul", "Trump", "None of the above"],
                    questions[2]: ["Modi", "Rahul", "Trump", "None of the above"],
                    questions[3]: ["Modi", "Rahul", "Trump", "None of the above"],
                    questions[4]: ["Modi", "Rahul", "Trump", "None of the above"],
                    questions[5]: ["Modi", "Rahul", "Trump", "None of the above"],
                    questions[6]: ["Modi", "Rahul", "Trump", "None of the above"],
                    questions[7]: ["Modi", "Rahul", "Trump", "None of the above"],
                    questions[8]: ["Modi", "Rahul", "Trump", "None of the above"],
                    questions[9]: ["Modi", "Rahul", "Trump", "None of the above"]}
```

```python
answer_of = {questions[0]: 1,
            questions[1]: 4,
            questions[2]: 3,
            questions[3]: 4,
            questions[4]: 4,
            questions[5]: 4,
            questions[6]: 4,
            questions[7]: 4,
            questions[8]: 4,
            questions[9]: 4}
```

```python
remarks = {0: 'work hard', 1: 'avg', 2: 'okok', 3: 'good', 4: 'very good', 5: 'Excellent'}
```

```python
columns = ['Name', 'Score', 'Remark']
df = pd.DataFrame(columns = columns)
```

```python
def remark(score: int):
  return remarks.get(score)
```

```python
def score(answers):
  final_score = 0
  for q in answers:
    if answers.get(q) == answer_of.get(q):
      final_score += 1
  return final_score
```

```python
def display_questions():
  answers = {}
  q = np.random.choice(questions, size=5, replace=False)
  for i in q:
    print(i)
    op = 0
    for j in questions_options.get(i):
      op+=1
      print(f'{op}. {j}')
    while True:
      try:
        op = int(input())
        if op > 4 or op < 1:
          print("wrong number, select again!")
        else:
          answers[i] = op
          break
      except:
        print("wrong input, select again!")
  
  return score(answers)
```

```python
def process():
  name = ''
  print('enter your name')
  while True:
    try:
      name = str(input())
      if name == "":
        print("enter valid name, cannot be empty")
        pass
      else:
        break
    except:
      print("wrong selection. select correct option")

  return name, display_questions()
```

```python
def repeat():
  while True:
    name, final_score = process()
    if final_score < 2:
      print('score is too low, try again!')
      pass
    else:
      global df
      new_row = {'Name': name,
                'Score': final_score,
                'Remark': remark(final_score)}
      df = df.append(new_row, ignore_index = True)
      break

def start():
  while True:
    print("select option:")
    print("1. Admin")
    print("2. Student")
    print("3. Exit")
    
    op = 0
    while True:
      try:
        op = int(input())
        if op > 3 or op < 1:
          print("wrong number, select again!")
        else:
          break
      except:
        print("wrong selection. select correct option")
    
    if op == 1:
      global df
      print(df)
    elif op == 2:
      repeat()
    elif op == 3:
      break
```

```python
start()
```
