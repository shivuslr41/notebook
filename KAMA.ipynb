```python
import pandas as pd
!pip install yfinance
import yfinance as yf
```

```python
data=pd.DataFrame()
data = yf.download("irctc.ns", period='max')
df2 = data[-2:-1]
df1 = data[-1:]
data.drop(data.tail(2).index,inplace=True)
```

```python
!pip install NitroFE
from NitroFE import KaufmanAdaptiveMovingAverage
```

```python
ob = KaufmanAdaptiveMovingAverage(kaufman_efficiency_lookback_period=21, kaufman_efficiency_min_periods=1, fast_ema_span=2, slow_ema_span=30)
res = ob.fit(data['Close'], first_fit=True)
print(res)
```

```python
import pickle
```

```python
file_to_store = open("pickles/stored_object.pickle", "wb")
pickle.dump(ob, file_to_store)
file_to_store.close()
```

```python
file_to_read = open("stored_object.pickle", "rb")
loaded_object = pickle.load(file_to_read)
```

```python
new_res=loaded_object.fit(df2['Close'],first_fit=False)
```

```python
concatenated_result=pd.concat([res,new_res],axis=0).reset_index(drop=True)
concatenated_result
```
